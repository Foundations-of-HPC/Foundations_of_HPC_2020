```python
import numpy as np
numproc=np.array([1, 12, 24, 48])
```

```python
numproc
```

```python
timeesec=np.array([2.90333333333333, 0.616666666666667, 0.686666666666667, 1.33333333333333])
t1=2.90333333333333
Spus=t1/timeesec
```

```python
Spus
```

```python
timeinter=np.array([2.62981359, 0.22037794, 0.11908735, 0.11482613])
t1i=2.62981359
Spis=t1i/timeinter
```

```python
Spis 
```

```python
import numpy as np
import matplotlib.pyplot as plt

plt.plot(numproc, Spus, 'r--', label="usr")
plt.plot(numproc, Spis, 'b--', label="int")
plt.xlabel('processors numbers')
plt.ylabel('speedup')
#plt.text(0,.00020, r'$N=2000$')
plt.suptitle('strong scalability per N=10^11')
plt.legend(loc='best')
plt.savefig('N=100miliardos.png')
plt.show()
```
